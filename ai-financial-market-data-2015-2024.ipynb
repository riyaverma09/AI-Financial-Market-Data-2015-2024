# High-Fidelity Synthetic Financial Market Data for Leading AI Models – Gemini (Google), ChatGPT (OpenAI), and LLaMA (Meta)

This dataset provides a synthetic, day-by-day record of financial market activities for companies leading in Artificial Intelligence (AI). It captures key financial metrics and major events that can influence stock performance, such as the launch of LLaMA by Meta, GPT by OpenAI, and Gemini by Google. The dataset also tracks how much these companies invest in R&D for their AI products and services, as well as the revenue generated from them. Covering the period from January 1, 2015, to December 31, 2024, it includes detailed information on three major AI-driven organizations: OpenAI, Google, and Meta.

The dataset is available in CSV format and will be analyzed using Pandas DataFrames. This analysis can be particularly valuable for professionals and researchers in the finance, investment, and stock market domains, providing insights into the financial trends of leading AI companies.

# How much amount the companies spent on R & D ?

```python
df.head()
```

```python
print("Company's spending for R & D in $Bn")

RD = df.groupby('Company')['R&D_Spending_USD_Mn'].sum()/1000

RD
```

```python
# Draw a Bar Plot to show the amount spent on R & D by the companies

plt.bar(RD.index, RD.values, color = ['cyan', 'black', 'magenta'])

plt.title( "R&D Spending by the companies")
plt.xlabel("Company")
plt.ylabel("Amount in USD_$Bn")

plt.show()
```

# **Revenue Earned by the companies**

```python
df.head()
```

```python
print("Company's AI_Revenue_USD_Bn :")

rev = df.groupby('Company')['AI_Revenue_USD_Mn'].sum()/1000

rev
```

```python
# Draw a Bar Plot to show the Revenues of the companies

plt.bar(rev.index, rev.values, color = ['Purple', 'Teal', 'Red'], width = 0.2)

plt.title( "Revenue earned by the companies")
plt.xlabel("Company")
plt.ylabel("Amount in USD_$Bn")

plt.show()
```

```python
# Bar plots to show expenditure & revenue of the companies

plt.figure(figsize = (11,4))

plt.subplot(1,2,1)

plt.bar(RD.index, RD.values, color = ['Purple', 'Teal', 'Red'])

plt.title( "R&D Spending by the companies")
plt.xlabel("Company")
plt.ylabel("Amount in USD_$Bn")

plt.subplot(1,2,2)

plt.bar(rev.index, rev.values, color = ['Purple', 'Teal', 'Red'], width = 0.2)

plt.title( "Revenue earned by the companies")
plt.xlabel("Company")
plt.ylabel("Amount in USD_$Bn")

plt.show()
```

# Expenditure vs Revenue year-by-year

```python
df.head()
```

```python
spend = df.groupby('Year')['R&D_Spending_USD_Mn'].sum()

spend
```

```python
# Showing the Amounnt spent on R & D

plt.plot(spend.index, spend.values, color = 'r')

plt.title("Combined R&D Spending Year-by-Year")
plt.xlabel("Year")
plt.ylabel("Amount in USD_$Mn")

plt.show()
```

```python
revenue = df.groupby('Year')['AI_Revenue_USD_Mn'].sum()

revenue
```

```python
# Showing the Revenue earned 

plt.plot( revenue.index, revenue.values, color = 'g')

plt.title( "Combined Revenue Earned Year-by-Year")
plt.xlabel("Year")
plt.ylabel("Amount in USD_$Mn")

plt.show()
```

```python
plt.plot(spend.index, spend.values, color = 'r')
plt.plot( revenue.index, revenue.values, color = 'g')

plt.title( "Combined Expenditure vs Revenue Year-by-Year", fontsize = 12)
plt.xlabel("Year")
plt.ylabel("Amount in USD_$Mn")

plt.legend(['Expenditure', 'Revenue'])

plt.show()
```
